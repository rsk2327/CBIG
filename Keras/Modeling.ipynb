```python
def preprocess(img):
    
    img = img + 15000
    
    shape = img.shape
    img = img.reshape((shape[0]*shape[1],))
    img = img.tolist()
    img = np.array([round(min(x,16384)*(255.0/16384.0)) for x in img])
    img = img.reshape(shape)
    img =np.uint8(img)
    
    return(img)
    
    
```
